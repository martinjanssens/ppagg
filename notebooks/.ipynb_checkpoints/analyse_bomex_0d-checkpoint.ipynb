```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import netCDF4 as nc
```

```python
rp = '/Users/martinjanssens/Documents/Wageningen/Patterns-in-satellite-images/BOMEXStability'
lps = [rp+'/bomex200_e12_small',
       rp+'/bomex200_e12'
      ]
sp = rp+'/comparison'

labs = [
       r'12.8 km',
       r'102.4 km',
       ]

ls = ['-',
      '--',
      ]

```

```python
fig = plt.figure(figsize=(5,5)); ax = plt.gca()
for i in range(len(lps)):
    ds0 = nc.Dataset(lps[i]+'/tmser.001.nc')
    time = ds0['time'][:].data / 3600
    zi = ds0['zi'][:].data
    
    ax.plot(time, zi, label=labs[i], linestyle=ls[i], c='k')
    ax.set_xlabel('Time [hr]',fontsize=12)
    ax.set_ylabel('Inversion height [m]',fontsize=12)
plt.legend(title='Domain size:', bbox_to_anchor=(1,1),loc='best', frameon=False)
sns.despine(offset=5)
plt.savefig(sp+'/inversion_height.pdf',bbox_inches='tight')
```
